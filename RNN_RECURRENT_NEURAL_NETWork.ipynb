```python
df=pd.read_csv("drive/MyDrive/data.csv")

for i in range(1000,5842):
  df.drop(i,inplace=True)

df['Sentiment']=df['Sentiment'].replace({'positive':2,'negative':0,'neutral':1})
pd.set_option('future.no_silent_downcasting', True)
```

```python
df['Sentence'] = df['Sentence'].str.lower()
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
```

```python
vocab = set([word for phrase in df['Sentence'] for word in phrase])
word_to_idx = {word: idx for idx, word in enumerate(vocab, 1)}
```

```python
vocab
```

```python
word_to_idx
```

```python
def encode_phrase(phrase):
  return [word_to_idx[word] for word in phrase]

train_data['Sentence'] = train_data['Sentence'].apply(encode_phrase)
test_data['Sentence'] = test_data['Sentence'].apply(encode_phrase)
```

```python
train_data
```

```python
# Padding sequences
max_length = max(df['Sentence'].apply(len))

def pad_sequence(seq, max_length):
    return seq + [0] * (max_length - len(seq))

train_data['Sentence'] = train_data['Sentence'].apply(lambda x: pad_sequence(x, max_length))
test_data['Sentence'] = test_data['Sentence'].apply(lambda x: pad_sequence(x, max_length))
```

```python
#Create Dataset and DataLoader
class SentimentDataset(Dataset):
    def __init__(self, data):
        self.Sentence = data['Sentence'].values
        self.Sentiment = data['Sentiment'].values

    def __len__(self):
        return len(self.Sentence)

    def __getitem__(self, idx):
        Sentences = self.Sentence[idx]
        Sentiments = self.Sentiment[idx]
        return torch.tensor(Sentences, dtype=torch.long), torch.tensor(Sentiments, dtype=torch.long)

train_dataset = SentimentDataset(train_data)
test_dataset = SentimentDataset(test_data)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
train_dataset
```

```python
class SentimentRNN(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size, output_size):
        super(SentimentRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_size)
        self.rnn = nn.RNN(embed_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.embedding(x)
        h0 = torch.zeros(1, x.size(0), hidden_size).to(x.device)
        out, _ = self.rnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out

vocab_size = len(vocab) + 1
embed_size = 64
hidden_size = 64
output_size = 3  # For binary classification
model = SentimentRNN(vocab_size, embed_size, hidden_size, output_size)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.005)

# Training loop
num_epochs = 50
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for Sentence, Sentiment in train_loader:
        outputs = model(Sentence)
        loss = criterion(outputs,Sentiment)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    if(epoch%10==0):
      print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss / len(train_loader):.4f}')
```

```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for Sentence, Sentiment in test_loader:
        outputs = model(Sentence)
        _, predicted = torch.max(outputs.data, 1)
        total += Sentiment.size(0)
        correct += (predicted == Sentiment).sum().item()

accuracy = 100 * correct / total
print(f'Accuracy: {accuracy:.2f}%')
```
